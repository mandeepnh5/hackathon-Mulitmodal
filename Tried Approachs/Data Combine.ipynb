```python
import os
import pandas as pd

# Assuming your current dataset is stored in a dataframe `df` and you want to include the following columns:
# - UserScreenName, UserName, Timestamp, Text, Embedded_text, interaction_score
def calculate_engagement_score(row):
    weight_replies = 1
    weight_retweets = 2
    weight_likes = 0.5

    engagement_score = (row['replies_count'] * weight_replies +
                        row['retweets_count'] * weight_retweets +
                        row['likes_count'] * weight_likes)
    
    return engagement_score

def process_bitcoins_tweets():
    # Path to the folder containing coin datasets
    base_folder = r"C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\train\tweets"  # Change this to your base folder path
    
    # List all folders in the base folder
    for coin_folder in os.listdir(base_folder):
        coin_folder_path = os.path.join(base_folder, coin_folder)
        
        # Make sure we're working with a folder
        if os.path.isdir(coin_folder_path):
            merged_tweets = []  # To store DataFrames from different CSV files
            
            # List all CSV files in the coin folder
            for file in os.listdir(coin_folder_path):
                if file.endswith(".csv"):
                    file_path = os.path.join(coin_folder_path, file)
                    df = pd.read_csv(file_path)
                    
                    # Select relevant columns and calculate engagement score
                    df['interaction_score'] = df.apply(calculate_engagement_score, axis=1)
                    relevant_columns = ['created_at', 'username', 'tweet', 'language' , 'cashtags', 'hashtags' , 'interaction_score']
                    
                    # Extract only the relevant columns
                    df_relevant = df[relevant_columns]
                    
                    # Add the date back as a column based on the filename
                    date = file.replace('.csv', '')
                    df_relevant['date'] = date
                    
                    # Append the DataFrame to the list
                    merged_tweets.append(df_relevant)
            
            # Concatenate all tweets for the coin into one DataFrame
            if merged_tweets:
                merged_df = pd.concat(merged_tweets, ignore_index=True)
                
                # Save the merged DataFrame to a new CSV file in the same folder
                output_file = os.path.join(coin_folder_path, f'{coin_folder}_merged_tweets.csv')
                merged_df.to_csv(output_file, index=False)
                print(f'Merged CSV saved for {coin_folder} at {output_file}')
            else:
                print(f'No CSV files found for {coin_folder}')
                
# Call the main function to process all coin tweet data
process_bitcoins_tweets()
```

```python
import pandas as pd
import os

# Define the base directory containing the dataset
base_directory = r"C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\test\tweets"  # Change this to your actual base directory
# Get a list of all Bitcoin folders
bitcoin_folders = os.listdir(base_directory)
desired_columns = [
    'created_at', 
    'username', 
    'tweet', 
    'language', 
    'cashtags', 
    'hashtags',   
    'replies_count', 
    'retweets_count', 
    'likes_count'
]
# Iterate over each Bitcoin folder
for bitcoin in bitcoin_folders:
    bitcoin_path = os.path.join(base_directory, bitcoin)
    
    # Check if the path is a directory
    if os.path.isdir(bitcoin_path):
        combined_tweets = []  # List to store DataFrames
        
        # Iterate through all the CSV files in the current Bitcoin folder
        for filename in os.listdir(bitcoin_path):
            if filename.endswith('.csv'):
                file_path = os.path.join(bitcoin_path, filename)
                
                try:
                    # Read the CSV file
                    df = pd.read_csv(file_path)
                    df = df[desired_columns]
                    df['interaction_score'] = (0.50 * df['replies_count'] + 
                                                0.75 * df['retweets_count'] + 
                                                0.25 * df['likes_count'])
                    df = df[['created_at','tweet','language','cashtags','hashtags','interaction_score']]
                    # Check if the DataFrame is empty
                    if df.empty:
                        print(f"Warning: The file {file_path} is empty and will be skipped.")
                        continue
                    df.dropna(subset=['tweet', 'created_at'], inplace=True)

                    # Remove rows where 'tweet' or 'date' are empty lists
                    df = df[~df['tweet'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
                    df = df[~df['created_at'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
                    
                    # Add a new column for the date
                    # df['date'] = filename[:-4]  # Remove .csv extension to get the date
                    combined_tweets.append(df)
                
                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")

        # Combine all DataFrames for this Bitcoin
        if combined_tweets:
            final_tweets = pd.concat(combined_tweets, ignore_index=True)
            # Save to a new CSV file
            output_file = os.path.join(base_directory, f"TWEETS/{bitcoin}_tweet.csv")
            final_tweets.to_csv(output_file, index=False)
            print(f"Combined tweets saved to {output_file}")
        else:
            print(f"No valid tweets found for {bitcoin}.")

print("Processing complete.")
##############################################
```

```python
import os
import pandas as pd
import numpy as np

# Directory paths
base_dir = r'C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\train\price_data'
output_dir = r'C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\train\price_data\NEW Features'

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Define a function to process each file and add features
def process_file(file_path, output_path):
    # Load the data
    data = pd.read_csv(file_path)
    
    # Ensure 'datetime' is in datetime format
    data['datetime'] = pd.to_datetime(data['datetime'])
    
    # Sort by date
    data = data.sort_values('datetime')

    ### Moving Averages ###
    data['MA_7_open'] = data['open_x'].rolling(window=7).mean()
    data['MA_14_open'] = data['open_x'].rolling(window=14).mean()
    data['MA_30_open'] = data['open_x'].rolling(window=30).mean()

    data['MA_7_close'] = data['close_x'].rolling(window=7).mean()
    data['MA_14_close'] = data['close_x'].rolling(window=14).mean()
    data['MA_30_close'] = data['close_x'].rolling(window=30).mean()

    data['MA_7_high'] = data['high_x'].rolling(window=7).mean()
    data['MA_14_high'] = data['high_x'].rolling(window=14).mean()
    data['MA_30_high'] = data['high_x'].rolling(window=30).mean()

    data['MA_7_low'] = data['low_x'].rolling(window=7).mean()
    data['MA_14_low'] = data['low_x'].rolling(window=14).mean()
    data['MA_30_low'] = data['low_x'].rolling(window=30).mean()

    ### Price Differences ###
    data['price_diff_close_open'] = data['close_x'] - data['open_x']
    data['price_diff_high_low'] = data['high_x'] - data['low_x']

    ### Trading Volume Ratios ###
    data['volume_ratio'] = data['volumeto_x'] / data['volumefrom_x']

    ### Sentiment Impact ###
    data['MA_7_sentiment'] = data['sentiment'].rolling(window=7).mean()
    data['MA_14_sentiment'] = data['sentiment'].rolling(window=14).mean()

    ### Lag Features ###
    data['lag_1_close'] = data['close_x'].shift(1)
    data['lag_7_close'] = data['close_x'].shift(7)
    data['lag_14_close'] = data['close_x'].shift(14)

    ### Handling Missing Values ###
    data.dropna(inplace=True)

    # Save the processed data to the output folder
    data.to_csv(output_path, index=False)

# Process all CSV files in the price_data folder
for file_name in os.listdir(base_dir):
    if file_name.endswith('.csv'):
        file_path = os.path.join(base_dir, file_name)
        output_path = os.path.join(output_dir, file_name)
        process_file(file_path, output_path)

print("Feature extraction complete. Files saved to the TWEET folder.")
```
