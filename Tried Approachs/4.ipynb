```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
data = pd.read_csv(r"C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\train\price_data\Feautures\AGI_merged.csv", parse_dates=['datetime'])

# Convert 'datetime' to datetime and set it as the index
data['datetimedatetime'] = pd.to_datetime(data['datetime'])  
data.set_index('datetimedatetime', inplace=True)

# Use 'close_x' as the target variable
target = data['close_x']

# Convert the datetime index into numeric values (for regression purposes)
data['numeric_date'] = (data.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')  # Days since epoch
numeric_dates = data['numeric_date'].values.reshape(-1, 1)

# Create Polynomial (Quadratic) Features for the numeric date
poly = PolynomialFeatures(degree=2)
numeric_dates_quad = poly.fit_transform(numeric_dates)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(numeric_dates_quad, target, test_size=0.2, random_state=42)

# Standardize the date feature
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and fit the quadratic regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Evaluate the model
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)

print(f"Train MSE: {mse_train}")
print(f"Test MSE: {mse_test}")
print(f"Train R^2: {r2_train}")
print(f"Test R^2: {r2_test}")

# Generate predictions across the entire dataset for the regression curve
X_scaled = scaler.transform(numeric_dates_quad)  # Scale the entire feature set
predicted_values = model.predict(X_scaled)

# Plot the actual vs predicted values with the quadratic regression curve
plt.figure(figsize=(12,6))
plt.plot(data.index, target, label='Actual Close_x', color='blue', lw=2)
plt.plot(data.index, predicted_values, label='Predicted Close_x (Quadratic Regression)', color='red', linestyle='--', lw=2)
plt.xlabel('Date')
plt.ylabel('Close_x Price')
plt.title('Quadratic Regression - Actual vs Predicted Close_x Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression

# Load the data (training data)
data = pd.read_csv(r"C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\train\price_data\Feautures\AGI_merged.csv", parse_dates=['datetime'])

# Convert 'datetime' to datetime and set it as the index
data['datetimedatetime'] = pd.to_datetime(data['datetime'])  
data.set_index('datetimedatetime', inplace=True)

# Use 'close_x' as the target variable
target = data['close_x']

# Convert the datetime index into numeric values (for regression purposes)
data['numeric_date'] = (data.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')  # Days since epoch
numeric_dates = data['numeric_date'].values.reshape(-1, 1)

# Create Polynomial (Quadratic) Features for the numeric date
poly = PolynomialFeatures(degree=2)
numeric_dates_quad = poly.fit_transform(numeric_dates)

# Create an 80:20 train-validation split using the index
train_size = int(len(numeric_dates_quad) * 0.8)
X_train, X_test = numeric_dates_quad[:train_size], numeric_dates_quad[train_size:]
y_train, y_test = target[:train_size], target[train_size:]

# Standardize the date feature
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create and fit the quadratic regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Load the test data (dates from the new file)
test_dates_df = pd.read_csv(r"C:\Users\gsanj\Downloads\public_dataset_updated.tar[1]\public_dataset\test\price_data\AGI.csv")

# Convert the 'date' column to datetime (automatically infers format)
test_dates_df['datetime'] = pd.to_datetime(test_dates_df['datetime'])

# Convert the dates into numeric form (days since epoch)
test_dates_df['numeric_date'] = (test_dates_df['datetime'] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')

# Create Polynomial (Quadratic) Features for the new test dates
numeric_test_dates = test_dates_df['numeric_date'].values.reshape(-1, 1)
numeric_test_dates_quad = poly.transform(numeric_test_dates)

# Scale the test dates using the same scaler fitted on training data
numeric_test_dates_scaled = scaler.transform(numeric_test_dates_quad)

# Make predictions for the new test dates
predicted_test_values = model.predict(numeric_test_dates_scaled)

# Display the predictions along with the test dates
test_dates_df['Predicted_Close_x'] = predicted_test_values
print(test_dates_df[['datetime', 'Predicted_Close_x']])

# Generate predictions across the entire dataset for the regression curve
X_scaled = scaler.transform(numeric_dates_quad)  # Scale the entire feature set
predicted_values = model.predict(X_scaled)

# Plot the actual vs predicted values with the quadratic regression curve
plt.figure(figsize=(12, 6))
plt.plot(data.index, target, label='Actual Close_x', color='blue', lw=2, alpha=0.7)
plt.plot(data.index, predicted_values, label='Predicted Close_x (Quadratic Regression)', color='red', linestyle='--', lw=2, alpha=0.7)

# Mark the predicted points on the test set
test_dates = data.index[train_size:]
plt.scatter(test_dates, y_test_pred, color='orange', label='Predicted Points (Test Set)', zorder=5)

# Plot the predictions for the new test dates
plt.scatter(test_dates_df['datetime'], predicted_test_values, color='green', label='Predicted Points (New Test Dates)', zorder=5)

# Plot formatting
plt.xlabel('Date')
plt.ylabel('Close_x Price')
plt.title('Quadratic Regression - Actual vs Predicted Close_x Prices')
plt.legend()
plt.grid(True)
plt.show()
```
